```python

import sys, pathlib, os

ROOT = pathlib.Path.cwd()  
if not (ROOT / "src").exists():
   
    nb_dir = pathlib.Path(__file__).resolve().parent if "__file__" in globals() else pathlib.Path.cwd()
    # walk up until we find src/
    cur = nb_dir
    while cur != cur.parent and not (cur / "src").exists():
        cur = cur.parent
    os.chdir(cur)
    ROOT = cur


sys.path[:0] = [str(ROOT), str(ROOT / "src")]

print("Working dir:", pathlib.Path.cwd())
print("Has src?:", (ROOT / "src").exists())

from src.config import DATA_PATH, TOP_N
from src.yelp_pipeline import (
    load_default_csv, build_synonyms, frequency_after_merge,
    select_top_by_cosine, sentiment_score, recommend_top3
)

import pandas as pd
from pathlib import Path
```

```python
df = load_default_csv(DATA_PATH)
print("CSV:", DATA_PATH)
print("Rows:", len(df), "| Unique restaurants:", df["restaurant"].nunique())
df.head(3)
```

```python
syn = build_synonyms()
freq = frequency_after_merge(df, syn)
freq.head(10)
```

```python
selected = select_top_by_cosine(df, syn, top_n=TOP_N)
selected[["restaurant","cosine_similarity"]].head(10)
```

```python
selected_sent = selected.copy()
selected_sent["sentiment"] = selected_sent["review"].apply(sentiment_score)
top3 = recommend_top3(selected)
top3
```

```python
def save(path, frame):
    p = Path(path)
    if p.parent.exists() or str(p.parent) == '.':
        frame.to_csv(p, index=False)
        print("Saved", p)
    else:
        frame.to_csv(p.name, index=False)
        print("Folder missing; saved", p.name)

save("outputs/frequency_table.csv", freq)
save("outputs/top200_reviews.csv", selected)
save("outputs/sentiment_top200.csv", selected_sent)
save("outputs/recommendations_top3.csv", top3)
```
